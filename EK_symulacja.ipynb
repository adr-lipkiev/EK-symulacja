```python
import pandas as pd
import plotly.graph_objects as go
```

```python
url_dochody = 'https://raw.githubusercontent.com/adr-lipkiev/EK-symulacja/refs/heads/main/dochody.csv'
url_wydatki = 'https://raw.githubusercontent.com/adr-lipkiev/EK-symulacja/refs/heads/main/wydatki.csv'
url_obliczenia = 'https://raw.githubusercontent.com/adr-lipkiev/EK-symulacja/refs/heads/main/obliczenia.csv'
```

```python
dochody = pd.read_csv(url_dochody, sep=';', decimal=',', na_values = ['(.)'], encoding='utf-8-sig')
wydatki = pd.read_csv(url_wydatki, sep=';', decimal=',', encoding='utf-8-sig')
obliczenia = pd.read_csv(url_obliczenia, sep=';', decimal=',', encoding='utf-8-sig')
```

###Sprawdzenie poprawności pobranych danych

```python
dochody = dochody.drop('Grupa społeczno-ekonomiczna (BGD)', axis=1)
```

```python
dochody
```

```python
wydatki = wydatki.dropna(how = 'all')
```

```python
wydatki
```

```python
obliczenia
```

#Obliczenia

##Stopa oszczędności

```python
obliczenia_dochod = obliczenia.iloc[0, 1:]
obliczenia_wydatki = obliczenia.iloc[1, 1:]
```

```python
stopa_oszczednosci = ((obliczenia_dochod - obliczenia_wydatki) / obliczenia_dochod) * 100
stopa_oszczednosci.index = stopa_oszczednosci.index.astype(int)
stopa_oszczednosci = stopa_oszczednosci.astype(float)
stopa_oszczednosci = stopa_oszczednosci.round(2)
```

```python
stopa_oszczednosci
```

```python
wykres1 = go.Figure()

przedzialy = [
    (2000, 2005, "20-25 lat", "rgba(255, 200, 200, 0.2)"),
    (2005, 2010, "25-30 lat", "rgba(255, 230, 150, 0.2)"),
    (2010, 2015, "30-35 lat", "rgba(255, 255, 150, 0.2)"),
    (2015, 2020, "35-40 lat", "rgba(200, 255, 200, 0.2)"),
    (2020, 2024, "40-45 lat", "rgba(180, 220, 255, 0.2)"),
]

for poczatek, koniec, etap, kolor in przedzialy:
  wykres1.add_vrect(
      x0 = poczatek,
      x1 = koniec,
      fillcolor = kolor,
      layer = 'below',
      line_width = 0,
      annotation_text = etap,
      annotation_position = "top left",
      annotation_font_size = 14,
      annotation_font_color = "gray"
  )

wykres1.add_trace(go.Scatter(
  x = stopa_oszczednosci.index,
  y = stopa_oszczednosci.values,
  mode = 'lines',
  line = dict(
      color = 'rgb(31, 119, 180)',
      width = 3
  ),
  showlegend = False
))

wykres1.update_layout(
    title = dict(
        text = 'Stopa oszczędności [%]',
        font = dict(size=24, color='black'),
        x = 0.5,
        xanchor = 'center'
    ),
    xaxis = dict(
        title = 'Rok',
        titlefont = dict(size=18),
        tickfont = dict(size=14),
        showgrid = False,
        showline = True,
        linewidth = 2,
        linecolor = 'black'
    ),
    yaxis = dict(
        title = 'Stopa oszczędności [%]',
        titlefont = dict(size=18),
        tickfont = dict(size=14),
        showgrid = True,
        gridcolor = 'lightgray',
        showline = True,
        linewidth = 2,
        linecolor = 'black'
    ),
    autosize = True,
    plot_bgcolor = 'white',
    paper_bgcolor = 'white',
    margin = dict(l=80, r=40, t=80, b=60)
)

wykres1.show()

wykres1.write_html('wykres_stopa_oszczednosci.html')
from google.colab import files
files.download('wykres_stopa_oszczednosci.html')
```

##WYDATKI

###Przyrost abosolutny łańcuchowy

```python
przyrost_abosolutny_l = wydatki.iloc[:, 1:].diff(axis=1)
przyrost_abosolutny_l.insert(0, 'Kategoria', wydatki['Rodzaj rozchodów'].values)
przyrost_abosolutny_l.columns = ['Kategoria'] + [int(col) for col in przyrost_abosolutny_l.columns[1:]]
```

```python
wykres_dochody_lan = go.Figure()

przedzialy = [
    (2000, 2005, "20-25 lat", "rgba(255, 200, 200, 0.2)"),
    (2005, 2010, "25-30 lat", "rgba(255, 230, 150, 0.2)"),
    (2010, 2015, "30-35 lat", "rgba(255, 255, 150, 0.2)"),
    (2015, 2020, "35-40 lat", "rgba(200, 255, 200, 0.2)"),
    (2020, 2024, "40-45 lat", "rgba(180, 220, 255, 0.2)"),
]

for poczatek, koniec, etap, kolor in przedzialy:
  wykres_dochody_lan.add_vrect(
      x0 = poczatek,
      x1 = koniec,
      fillcolor = kolor,
      layer = 'below',
      line_width = 0,
      annotation_text = etap,
      annotation_position = "top left",
      annotation_font_size = 14,
      annotation_font_color = "gray"
  )

for i in range(len(przyrost_wzg_lan_dochody)):
  wykres_dochody_lan.add_trace(go.Scatter(
      x = przyrost_wzg_lan_dochody.columns[1:],
      y = przyrost_wzg_lan_dochody.iloc[i, 1:].values,
      mode = 'lines',
      name = przyrost_wzg_lan_dochody.iloc[i, 0],
      line = dict(width=2.5)
  ))

wykres_dochody_lan.update_layout(
    title = dict(
        text = 'Przyrost dochodów względny łańcuchowy [%]',
        font = dict(size=24, color='black'),
        x = 0.5,
        xanchor = 'center'
    ),
    xaxis = dict(
        title = 'Rok',
        titlefont = dict(size=18),
        tickfont = dict(size=14),
        showgrid = False,
        showline = True,
        linewidth = 2,
        linecolor = 'black'
    ),
    yaxis = dict(
        title = 'Przyrost [%]',
        titlefont = dict(size=18),
        tickfont = dict(size=14),
        showgrid = True,
        gridcolor = 'lightgray',
        showline = True,
        linewidth = 2,
        linecolor = 'black'
    ),
    autosize = True,
    plot_bgcolor = 'white',
    paper_bgcolor = 'white',
    margin = dict(l=80, r=40, t=80, b=100),
    legend = dict(
        font = dict(size=12),
        orientation = 'h',
        yanchor = 'top',
        y = -0.15,
        xanchor = 'center',
        x = 0.5
    )
)

wykres_dochody_lan.show()
```
