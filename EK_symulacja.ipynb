#Analiza wydatków oraz dochodów dla grupy osób pracujących na własny rachunek

```python
#DOCHODY

dochody_analiza = dochody.copy()
dochody_analiza['Wartość 2000'] = dochody_analiza['2000']
dochody_analiza['Wartość 2024'] = dochody_analiza['2024']
dochody_analiza['Zmiana'] = dochody_analiza['2024'] - dochody_analiza['2000']
dochody_analiza['Zmiana %'] = ((dochody_analiza['2024'] - dochody_analiza['2000']) / dochody_analiza['2000']) * 100

dochody_ranking = dochody_analiza[['Rodzaj przychodów', 'Wartość 2000', 'Wartość 2024', 'Zmiana', 'Zmiana %']].sort_values('Zmiana %', ascending=False)

print("="*80)
print("\n RANKING WZROSTU DOCHODÓW (2000 - 2024)\n")
print("="*80)

for index, row in dochody_ranking.iterrows():
    print(f"\n {row['Rodzaj przychodów']}")
    print(f"   2000: {row['Wartość 2000']:>10.2f}")
    print(f"   2024: {row['Wartość 2024']:>10.2f}")
    print(f"   Zmiana: {row['Zmiana']:>8.2f}  ({row['Zmiana %']:+.1f}%)")

#WYDATKI

wydatki_analiza = wydatki.copy()
wydatki_analiza['Wartość 2000'] = wydatki_analiza['2000']
wydatki_analiza['Wartość 2024'] = wydatki_analiza['2024']
wydatki_analiza['Zmiana'] = wydatki_analiza['2024'] - wydatki_analiza['2000']
wydatki_analiza['Zmiana %'] = ((wydatki_analiza['2024'] - wydatki_analiza['2000']) / wydatki_analiza['2000']) * 100

wydatki_ranking = wydatki_analiza[['Rodzaj rozchodów', 'Wartość 2000', 'Wartość 2024', 'Zmiana', 'Zmiana %']].sort_values('Zmiana %', ascending=False)

print("\n" + "="*80)
print("\n RANKING WZROSTU WYDATKÓW (2000 - 2024)\n")
print("="*80)

for index, row in wydatki_ranking.iterrows():
    print(f"\n {row['Rodzaj rozchodów']}")
    print(f"   2000: {row['Wartość 2000']:>10.2f}")
    print(f"   2024: {row['Wartość 2024']:>10.2f}")
    print(f"   Zmiana: {row['Zmiana']:>8.2f}  ({row['Zmiana %']:+.1f}%)")
```

```python
fig1 = go.Figure()

przedzialy = [
    (2000, 2005, "20-25 lat", "rgba(255, 200, 200, 0.2)"),
    (2005, 2010, "25-30 lat", "rgba(255, 230, 150, 0.2)"),
    (2010, 2015, "30-35 lat", "rgba(255, 255, 150, 0.2)"),
    (2015, 2020, "35-40 lat", "rgba(200, 255, 200, 0.2)"),
    (2020, 2024, "40-45 lat", "rgba(180, 220, 255, 0.2)"),
]

for start, koniec, wiek, kolor in przedzialy:
    fig1.add_vrect(
        x0=start,
        x1=koniec,
        fillcolor=kolor,
        layer="below",
        line_width=0,
        annotation_text=wiek,
        annotation_position="top left",
        annotation_font_size=11,
        annotation_font_color="gray"
    )

lata = [int(rok) for rok in dochody.columns[1:]]

for index, row in dochody.iterrows():
    kategoria = row['Rodzaj przychodów']
    wartosci = row[1:].values

    fig1.add_trace(go.Scatter(
        x=lata,
        y=wartosci,
        mode='lines+markers',
        name=kategoria,
        hovertemplate='<b>%{fullData.name}</b><br>' +
                      'Rok: %{x}<br>' +
                      'Wartość: %{y:.2f} zł<br>' +
                      '<extra></extra>'
    ))

fig1.update_layout(
    title='DOCHODY - Wartości bezwzględne (2000-2024)<br>',
    xaxis_title='Rok',
    yaxis_title='Wartość (zł)',
    hovermode='x unified',
    template='plotly_white',
    autosize=True,
    margin=dict(l=50, r=50, t=100, b=50),
    legend=dict(orientation="h", yanchor="bottom", y=-0.3, xanchor="center", x=0.5)
)

fig1.show()
```

```python
wydatki_rdr = wydatki[['Rodzaj rozchodów']].copy()
lata_kolumny = wydatki.columns[1:]

for index, row in wydatki.iterrows():
    for i in range(1, len(lata_kolumny)):
        rok_poprzedni = lata_kolumny[i-1]
        rok_obecny = lata_kolumny[i]
        wartosc_poprzednia = row[rok_poprzedni]
        wartosc_obecna = row[rok_obecny]

        if pd.notna(wartosc_poprzednia) and pd.notna(wartosc_obecna) and wartosc_poprzednia > 0:
            zmiana_proc = ((wartosc_obecna - wartosc_poprzednia) / wartosc_poprzednia) * 100
            wydatki_rdr.at[index, rok_obecny] = zmiana_proc
        else:
            wydatki_rdr.at[index, rok_obecny] = None

fig3 = go.Figure()

przedzialy = [
    (2000, 2005, "20-25 lat", "rgba(255, 200, 200, 0.2)"),
    (2005, 2010, "25-30 lat", "rgba(255, 230, 150, 0.2)"),
    (2010, 2015, "30-35 lat", "rgba(255, 255, 150, 0.2)"),
    (2015, 2020, "35-40 lat", "rgba(200, 255, 200, 0.2)"),
    (2020, 2024, "40-45 lat", "rgba(180, 220, 255, 0.2)"),
]

for start, koniec, wiek, kolor in przedzialy:
    fig3.add_vrect(x0=start, x1=koniec, fillcolor=kolor, layer="below", line_width=0,
                   annotation_text=wiek, annotation_position="top left",
                   annotation_font_size=11, annotation_font_color="gray")

lata = [int(rok) for rok in wydatki.columns[1:]]

for index, row in wydatki.iterrows():
    fig3.add_trace(go.Scatter(
        x=lata, y=row[1:].values, mode='lines+markers', name=row['Rodzaj rozchodów'],
        hovertemplate='<b>%{fullData.name}</b><br>Rok: %{x}<br>Wartość: %{y:.2f} zł<br><extra></extra>'
    ))

fig3.update_layout(
    title='WYDATKI - Wartości bezwzględne (2000-2024)',
    xaxis_title='Rok', yaxis_title='Wartość (zł)', hovermode='x unified',
    template='plotly_white', autosize=True,
    margin=dict(l=50, r=50, t=100, b=50),
    legend=dict(orientation="h", yanchor="bottom", y=-0.4, xanchor="center", x=0.5)
)

fig3.show()
```

```python
podstawowe = ['Żywność i napoje bezalkoholowe',
              'Użytkowanie mieszkania i nośniki energii',
              'Zdrowie',
              'Odzież i obuwie']

luksusowe = ['Napoje alkoholowe, wyroby tytoniowe i narkotyki',
             'Rekreacja i kultura',
             'Transport',
             'Łączność',
             'Edukacja',
             'Wyposażenie mieszkania i prowadzenie gospodarstwa domowego']

wydatki_podst = wydatki[wydatki['Rodzaj rozchodów'].isin(podstawowe)]
wydatki_luks = wydatki[wydatki['Rodzaj rozchodów'].isin(luksusowe)]

suma_podst = wydatki_podst.iloc[:, 1:].sum()
suma_luks = wydatki_luks.iloc[:, 1:].sum()

baza_podst = suma_podst.iloc[0]
baza_luks = suma_luks.iloc[0]

zmiana_podst = ((suma_podst - baza_podst) / baza_podst) * 100
zmiana_luks = ((suma_luks - baza_luks) / baza_luks) * 100

fig5 = go.Figure()

for start, koniec, wiek, kolor in przedzialy:
    fig5.add_vrect(x0=start, x1=koniec, fillcolor=kolor, layer="below", line_width=0,
                   annotation_text=wiek, annotation_position="top left",
                   annotation_font_size=11, annotation_font_color="gray")

lata = [int(rok) for rok in wydatki.columns[1:]]

fig5.add_trace(go.Scatter(x=lata, y=zmiana_podst.values, mode='lines+markers',
                          name='Dobra podstawowe', line=dict(width=3),
                          hovertemplate='<b>Dobra podstawowe</b><br>Rok: %{x}<br>Zmiana: %{y:+.1f}%<br><extra></extra>'))

fig5.add_trace(go.Scatter(x=lata, y=zmiana_luks.values, mode='lines+markers',
                          name='Dobra luksusowe', line=dict(width=3),
                          hovertemplate='<b>Dobra luksusowe</b><br>Rok: %{x}<br>Zmiana: %{y:+.1f}%<br><extra></extra>'))

fig5.add_hline(y=0, line_dash="dash", line_color="gray", line_width=2)

fig5.update_layout(
    title='WYDATKI - Dobra podstawowe vs luksusowe - zmiana % względem 2000',
    xaxis_title='Rok', yaxis_title='Zmiana względem 2000 (%)', hovermode='x unified',
    template='plotly_white', autosize=True,
    margin=dict(l=50, r=50, t=100, b=50),
    legend=dict(orientation="h", yanchor="bottom", y=-0.4, xanchor="center", x=0.5)
)

fig5.show()
```

```python
zmiana_rdr_podst = suma_podst.pct_change() * 100
zmiana_rdr_luks = suma_luks.pct_change() * 100

fig6 = go.Figure()

for start, koniec, wiek, kolor in przedzialy:
    fig6.add_vrect(x0=start, x1=koniec, fillcolor=kolor, layer="below", line_width=0,
                   annotation_text=wiek, annotation_position="top left",
                   annotation_font_size=11, annotation_font_color="gray")

lata = [int(rok) for rok in wydatki.columns[1:]]

fig6.add_trace(go.Scatter(x=lata, y=zmiana_rdr_podst.values, mode='lines+markers',
                          name='Dobra podstawowe', line=dict(width=3),
                          hovertemplate='<b>Dobra podstawowe</b><br>Rok: %{x}<br>Zmiana: %{y:+.1f}%<br><extra></extra>'))

fig6.add_trace(go.Scatter(x=lata, y=zmiana_rdr_luks.values, mode='lines+markers',
                          name='Dobra luksusowe', line=dict(width=3),
                          hovertemplate='<b>Dobra luksusowe</b><br>Rok: %{x}<br>Zmiana: %{y:+.1f}%<br><extra></extra>'))

fig6.add_hline(y=0, line_dash="dash", line_color="gray", line_width=2)

fig6.update_layout(
    title='WYDATKI - Dobra podstawowe vs luksusowe - zmiana % rok do roku',
    xaxis_title='Rok', yaxis_title='Zmiana rok-do-roku (%)', hovermode='x unified',
    template='plotly_white', autosize=True,
    margin=dict(l=50, r=50, t=100, b=50),
    legend=dict(orientation="h", yanchor="bottom", y=-0.4, xanchor="center", x=0.5)
)

fig6.show()
```
